```python
import pickle
import os
import glob
from IPython.display import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

for folder in glob.glob('./experiment_data/*'):
    print(folder)
    for file in glob.glob(folder+'/*.pkl'):
        infile = open(file,'rb')
        new_dict = pickle.load(infile)
        infile.close()
        best = new_dict['best']
        worst = new_dict['worst']
        for img in best:
            fig, ax = plt.subplots()
            plt.xticks([])
            plt.yticks([])
            filename = img[0]
            imag = mpimg.imread('./data/images/test/'+filename)
            plt.imshow(imag)
            plt.show()
            print('Given Captions:')
            for cap in img[3]:
                print(' '.join(cap))
            print('Generated Caption:')
            print(' '.join(img[2]))
            
        for img in worst:
            fig, ax = plt.subplots()
            
            filename = img[0]
            plt.xticks([])
            plt.yticks([])
            imag = mpimg.imread('./data/images/test/'+filename)
            plt.imshow(imag)
            
            plt.show()
            
            print('Given Captions:')
            for cap in img[3]:
                print(' '.join(cap))
            print('/n')
            print('Generated Caption:')
            print(' '.join(img[2]))
        
```
